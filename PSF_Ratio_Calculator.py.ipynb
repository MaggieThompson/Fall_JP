```python
def source_box(data,x,y,boxsize):
    if (x-boxsize)<0:
        x_low=0
    else:
        x_low=x-boxsize
    if (x+boxsize)>data.shape[0]:
        x_high=data.shape[0]
    else:
        x_high=(x+boxsize)
    if (y-boxsize)<0:
        y_low=0
    else:
        y_low=(y-boxsize)
    if (y+boxsize)>data.shape[1]:
        y_high=data.shape[1]
    else:
        y_high=(y+boxsize)
    return data[y_low:y_high, x_low:x_high]
        

def graph_psf_params(image_file, sources_file):
    #Part 1: download .fits file:
    hdu_list_image=fits.open(image_file)
    image_data=hdu_list_image[0].data
    #Part 2: download .txt file, convert to pixel coordinates:
    nearby_sources=np.genfromtxt(sources_file, skip_header=23, usecols=(0,1))
    sources_data=pd.DataFrame(nearby_sources, columns=('RA', 'DEC'))
    w=WCS(image_file)
    px,py=w.all_world2pix(sources_data['RA'],sources_data['DEC'],1, tolerance=200.0)
    index=(np.arange(1,len(px)+1))
    x,y=px[0],py[0]
    IMAGE_DATA=source_box(image_data,x,y,20.0)
    mpfit = gauss.gaussfit(IMAGE_DATA, returnmp=True)
    gaussianfit=gauss.twodgaussian(mpfit.params)
    plt.figure()
    plt.imshow(IMAGE_DATA, cmap='gray_r')
    plt.colorbar()
    plt.contour(gaussianfit(*np.indices(IMAGE_DATA.shape)))
    plt.title(image_file)
    plt.show()
    
#KOI's between 1.5-2.0
#KOI6130   
graph_psf_params('./Potential_Binaries/6130Ic-dc20140206UKJ.fits', './Potential_Binaries/6130Ic-dc20140206UKJ.src')   
#KOI5552
graph_psf_params('./Potential_Binaries/5552Ic-dc20140206UKJ.fits', './Potential_Binaries/5552Ic-dc20140206UKJ.src')
#KOI5997-Kepler Eclipsing Binary
graph_psf_params('./Potential_Binaries/5997Ic-dc20140206UKJ.fits', './Potential_Binaries/5997Ic-dc20140206UKJ.src')
#KOI53-Kepler Eclipsing Binary
#graph_psf_params('./Potential_Binaries/53Ic-dc20140206UKJ.fits', './Potential_Binaries/53Ic-dc20140206UKJ.src')
#KOI3954
graph_psf_params('./Potential_Binaries/3954Ic-dc20140206UKJ.fits', './Potential_Binaries/3954Ic-dc20140206UKJ.src')
#KOI5668
graph_psf_params('./Potential_Binaries/5668Ic-dc20140206UKJ.fits', './Potential_Binaries/5668Ic-dc20140206UKJ.src')
#KOI389
graph_psf_params('./Potential_Binaries/389Ic-dc20140206UKJ.fits', './Potential_Binaries/389Ic-dc20140206UKJ.src')
#KOI606-Kepler Eclipsing Binary
graph_psf_params('./Potential_Binaries/606Ic-dc20140206UKJ.fits', './Potential_Binaries/606Ic-dc20140206UKJ.src')
#KOI697
graph_psf_params('./Potential_Binaries/697Ic-dc20140206UKJ.fits', './Potential_Binaries/697Ic-dc20140206UKJ.src')
#KOI6132
graph_psf_params('./Potential_Binaries/6132Ic-dc20140206UKJ.fits', './Potential_Binaries/6132Ic-dc20140206UKJ.src')
#KOI4872
graph_psf_params('./Potential_Binaries/4872Ic-dc20140206UKJ.fits', './Potential_Binaries/4872Ic-dc20140206UKJ.src')
```
